```python
#goals = data_df.query("bullet_type == 1 or bullet_type == 4")
goals = data_df

goals['creation_datetime'] = pd.to_datetime(goals.creation_time)
goals['datetime'] = pd.to_datetime(goals.dt)
goals['completed_datetime'] = pd.to_datetime(goals.completed_time)
goals['date'] = goals['datetime'].dt.date
goals['creation_date'] = goals['creation_datetime'].dt.date
```

```python
task_columns = ['date','text','bullet_status']
tasks_df = goal_dates.query("bullet_type == 1")[task_columns]
```

```python
tasks_df.head()
```

```python
tasks_df['text'] = tasks_df['text'].str.replace("#","")
```

```python
tasks_df.groupby('text').count()
```

```python
# replace {old_name: new name}
duplicates_dict = {
    "Laundry":"Start a load of laundry",
    "Create a Finch_Backup_File":"Create a Finch Backup File",
    "Walk around the neighborhood":"Mindfulness Walk"
}
```

```python
tasks_df['text'] = tasks_df['text'].replace(duplicates_dict)
```

```python
task_names = tasks_df.groupby('text').count().index.to_list()
task_names
```
