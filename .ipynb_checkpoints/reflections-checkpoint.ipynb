```python
import pandas as pd
```

```python
data_location = 'C:/Users/kaley/OneDrive/Documents/Coding Stuff/Data Science/Mental Health Project/Sample Data/'
variables_data_location = data_location+'FinchExport_2025-03-01/Bullet.json'
```

```python
data_series = pd.read_json(variables_data_location, typ='series')
data_df = pd.json_normalize(data_series.data)
```

```python
data_df.head()
```

```python
cols = ['dt','creation_time','bullet_type','text']
reflections_df = data_df[cols].query("bullet_type == 2")
```

```python
reflections_df.head()
```

```python
reflections_df['datetime'] = pd.to_datetime(reflections_df.dt)
reflections_df['creation_datetime'] = pd.to_datetime(reflections_df.creation_time)
reflections_df['date'] = reflections_df['datetime'].dt.date
reflections_df['creation_date'] = reflections_df['creation_datetime'].dt.date
```

```python
reflections_df.head()
```

```python
reflections_df = reflections_df[['date','creation_date','datetime','creation_datetime','text']]
```

```python
reflections_df.head()
```

```python
reflections_df.sort_values('datetime', ascending=False)
```

* date: date (year, month, day) of the goals
* datetime: date and time the goals were added ( generally 1am because that's when the app created the new day's goals? )
* creation date: the date the reflection was written
* creation datetime: date and time the reflection was written
* text: the text of the reflection, ie. what I wrote

#### NOTE: creation_date may differ from date if reflection was written after midnight. The reflection is still considered part of the day before, but creation_datetime records when the reflection was created
