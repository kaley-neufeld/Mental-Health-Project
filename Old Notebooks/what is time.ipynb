```python
import pandas as pd
import matplotlib.pyplot as plt  
from datetime import datetime, timedelta, date
```

```python
data_location = 'C:/Users/kaley/OneDrive/Documents/Coding Stuff/Data Science/Mental Health Project/Sample Data/'
# C:\Users\kaley\OneDrive\Documents\Coding Stuff\Data Science\Mental Health Project\Sample Data\FinchExport_2024-09-23
variables_data_location = data_location+'FinchExport_2025-04-12/Bullet.json'
```

```python
data_series = pd.read_json(variables_data_location, typ='series')
data_df = pd.json_normalize(data_series.data)
```

```python
exvar_df = data_df = pd.json_normalize(data_series.data)[['dt','bullet_type','text','bullet_status']]
exvar_df = exvar_df.query("bullet_type == 1")
exvar_df = exvar_df.drop('bullet_type',axis=1)
```

```python
# convert to datetime
exvar_df['Record_Datetime'] = exvar_df.dt.map(lambda x: datetime.strptime(x, "%a, %d %b %Y %H:%M:%S"))
exvar_df['Record_Date']= exvar_df['Record_Datetime'].dt.date
exvar_df = exvar_df.drop(['dt'],axis=1)
```

```python
exvar_df['text'] = exvar_df['text'].str.replace("#","")
exvar_df.head()
```

```python
# replace {old_name: new name}
duplicates_dict = {
    "Laundry":"Start a load of laundry",
    "Create a Finch_Backup_File":"Create a Finch Backup File",
    "Walk around the neighborhood":"Mindfulness Walk"
}
```

```python
exvar_df['text'] = exvar_df['text'].replace(duplicates_dict)
```

```python
exvar_df = exvar_df.sort_values('Record_Date',ascending=False)
exvar_df.head()
```

```python
start_date = date(2025,3,29)
end_date = date(2025,4,11)
recent_df = exvar_df[(exvar_df['Record_Date'] > start_date) & (exvar_df['Record_Date'] < end_date)].sort_values('Record_Datetime')
```
